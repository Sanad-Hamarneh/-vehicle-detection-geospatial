```python
import os
from PIL import Image
import tifffile as tiff
import numpy as np

# Define the input and output directories
input_dir = r'C:\Users\USER\Desktop\png img labelimg'
output_dir = r'C:\Users\USER\Desktop\png img labelimg\converted'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Loop through each file in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.tif') or filename.endswith('.tiff'):
        try:
            # Construct the full file path
            input_path = os.path.join(input_dir, filename)
            output_path = os.path.join(output_dir, os.path.splitext(filename)[0] + '.png')

            # Open the TIFF file
            tiff_image = tiff.imread(input_path)

            # Convert the image to a PIL Image
            pil_image = Image.fromarray(np.uint8(tiff_image))

            # Save as PNG
            pil_image.save(output_path)

            print(f"Converted {filename} to PNG.")
        except Exception as e:
            print(f"An error occurred with file {filename}: {e}")
    else:
        print(f"Skipping non-TIFF file: {filename}")

print("Batch conversion completed!")
```

```python
import os
from PIL import Image
import tifffile as tiff
import numpy as np

# Define the input and output directories
input_dir = r'D:\for interns\no black 15%\1'
output_dir = r'D:\for interns\15% jepy\1'

# Create the output directory if it doesn't exist
os.makedirs(output_dir, exist_ok=True)

# Loop through each file in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.tif') or filename.endswith('.tiff'):
        try:
            # Construct the full file path
            input_path = os.path.join(input_dir, filename)
            output_path = os.path.join(output_dir, os.path.splitext(filename)[0] + '.jpeg')

            # Open the TIFF file
            tiff_image = tiff.imread(input_path)

            # Convert the image to a PIL Image
            pil_image = Image.fromarray(np.uint8(tiff_image))

            # Convert the image to RGB mode if necessary
            if pil_image.mode != 'RGB':
                pil_image = pil_image.convert('RGB')

            # Save as JPEG
            pil_image.save(output_path, 'JPEG')

            print(f"Converted {filename} to JPEG.")
        except Exception as e:
            print(f"An error occurred with file {filename}: {e}")
    else:
        print(f"Skipping non-TIFF file: {filename}")

print("Batch conversion completed!")
```
